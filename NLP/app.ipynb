```python
from nltk.tokenize import sent_tokenize
```

```python
from nltk.tokenize import word_tokenize
```

```python
words = word_tokenize(corpus)
print(words)
print(len(words))
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
words = wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
words = tokenizer.tokenize(corpus)
print(len(words))
```

# Stemming & Lemmatization

```python
from nltk.stem import PorterStemmer # RegexStemmer, Snowball Stemmer
```

```python
n_words = words[0:21]
n_words
```

```python
stemmer = PorterStemmer()
```

```python
for word in n_words:
    print(f"{word}-----> {stemmer.stem(word)}")
```

```python
from nltk.stem import SnowballStemmer # RegexStemmer, Snowball Stemmer
```

```python
snw_stemmer = SnowballStemmer('english')
```

```python
for word in n_words:
    print(f"{word}-----> {snw_stemmer.stem(word)}")
```

```python
from nltk.stem import WordNetLemmatizer 
```

```python
lemmarizer = WordNetLemmatizer()
```

```python
lemmarizer.lemmatize('going') 
```

```python
lemmarizer.lemmatize('going', pos='a') # pos parameter indicates the type of word/term in English eg verb, adverb , noun(default), pronoun
```

```python
# print("WORD   ----->    STEM    -----> LEMMA")
for word in n_words:
    print(f"| WORD: {word}| -----> || STEM:{snw_stemmer.stem(word)}    LEMMA: {lemmarizer.lemmatize(word)} ||")
```

```python
# print("WORD   ----->    STEM    -----> LEMMA")
for word in n_words:
    print(f"| WORD: {word}| -----> || STEM:{snw_stemmer.stem(word)}    LEMMA: {lemmarizer.lemmatize(word, pos='v')} ||")
```
